```python
from transformers import pipeline
pipe = pipeline("text-generation", model="Qwen/Qwen2.5-1.5B-Instruct")
```

```python
question1 = "What is an ACE?"
question2 = "What is a volley?"

prompt = "Answer the question\n" + \
"Question: "+ question1 + "\n" + \
"Answer: "
print(prompt)
```

```python
result = pipe(prompt, max_new_tokens = 50)
print(result)

#generated something meaningless answer...as didn't pass any context
```
