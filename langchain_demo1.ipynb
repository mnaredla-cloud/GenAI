```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
```

```python
pipe = pipeline("text-generation", model="Qwen/Qwen2.5-1.5B-Instruct", max_length = 512)
```

```python
#wrap the pipeline in langchain
#HuggingFacePipeline acts as a bridge between HuggingFace and Langchain

chat_model = HuggingFacePipeline(pipeline=pipe)
```

```python
response = chat_model.invoke("Explain LangChain in simple terms?")
```
