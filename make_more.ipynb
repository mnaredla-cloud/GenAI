```python
#For each name, convert string to list and add <S> at start and <E> at the end.
# chr= ['<S>', 'V', 'i', 'h', 'a', 'a', 'n', '<E>']

#for ch1,ch2 in zip(chr,chr[1:]):
#take input of chr and input of chr by neglecting 1st charector

for name in names:
  chr = ['<S>'] + list(name) + ['<E>']
  for ch1,ch2 in zip(chr,chr[1:]):
    print(ch1,ch2)
```

```python
#define a empty dictonary >>b
#tuple of ch1,ch2   >>('<S>', 'V')

b = {}
for name in names[0:1]:
  chr = ['<S>'] + list(name) + ['<E>']
  for ch1,ch2 in zip(chr,chr[1:]):
    biagram = (ch1,ch2)
    b[biagram] = b.get(biagram,0) + 1
```

```python
# Print output of dictonary for 1 input

print(b)
```

```python
#print unique charectors

chars = sorted(list(set(''.join(names))))
print(chars)
```

```python
#For printing index, value of the unique charectors

for i,s in enumerate(chars):
  print(i,s)
```

```python
#interchange index,value with value,index
#and dtype to dict with values:key

stoi = {s:i for i,s in enumerate(chars)}
```

```python
#add two more vales with it's index values at 26,27

stoi['<S>'] = 26
stoi['<E>'] = 27
```

```python
print(stoi)
```

```python
#index_1 =index[ch1]
#index_2 =index[ch2]
#T[index_1,index_2] = T[index_1,index_2] + 1
#T-Tensor

for W in names:
  chr = ['<S>'] + list(W) + ['<E>']
  for ch1,ch2 in zip(chr,chr[1:]):
    index_1 = stoi[ch1]
    index_2 = stoi[ch2]
    T[index_1,index_2] += 1
```
