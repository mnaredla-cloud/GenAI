```python
#Neural network approach

words = open('names.txt','r').read().splitlines()
```

```python
#Neural network approach

xs ,ys = [],[]


for w in words[0:1]:
  chs = ['.'] + list(w) + ['.']
  for ch1,ch2 in zip(chs,chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    xs.append(ix1)
    ys.append(ix2)


xs = torch.tensor(xs)
ys = torch.tensor(ys)
print(xs)
print(ys)
```

```python
import torch.nn.functional as F
xenc = F.one_hot(xs,num_classes=27).float()
xenc
```

```python
xenc.shape
```

```python
xenc.dtype
```

```python
#Create 27x27 neurons

W = torch.randn(27,27)
W
```

```python
# xenc (5,27)
# W (27,27)
# xenc x W

xenc @ W
```

```python
(xenc @ W).shape
```

```python
(xenc @ W)[3,13]
```

```python
logits = xenc @ W
counts = logits.exp()
probs = counts / counts.sum(1,keepdims=True)
probs
```
